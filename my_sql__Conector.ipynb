## Connection with iteration.

```python
import mysql.connector as c
con=c.connect(host="localhost", user="root",password="#######",
              database="db1")
cur=con.cursor()
cur.execute("select * from students")

x=cur.fetchmany(3)

for i in x:
    for j in i:
        print(j,end="\t")
    print()
con.close()
```

## Creating and edit Mysql table

```python
import mysql.connector as x

def myConnection():
    con=x.connect(host="localhost",user="root",password="#####",database="db4")
    return con

def display():
    con=myConnection()
    cur=con.cursor()
    cur.execute("select * from mobile")
    rec=cur.fetchall()
    for i in rec:
        for j in i:
            print(j,end="\t")
        print()
    con.close()

def save():
    con=myConnection()
    cur=con.cursor()
    mcode=int(input("enter mobile code: "))
    brand=input("enter brand name: ")
    price=int(input("enter price: "))
    
    cur.execute("insert into mobile values('%d','%s','%d')" % (mcode,brand,price))
    con.commit()
    print("data saved")
    con.close()
    
def search():
    con=myConnection()
    cur=con.cursor()
    brand=input("enter brand name to search: ")
    cur.execute("select * from mobile where brand='%s'" % (brand))
    x=cur.fetchall()
    if(cur.rowcount>0):
        for i in x:
            for j in i:
                print(j,end="\t")
            print()
    else:
        print("record not found")
    con.close()
    
def update():
    con=myConnection()
    cur=con.cursor()
    mcode=int(input("enter mobile code to update: "))
    price=int(input("enter new price: "))
    cur.execute("update mobile set price='%d' where mcode='%d'" % (price,mcode))
    con.commit()
    con.close()
    
    if(cur.rowcount>0):
        print("record updated")
    else:
        print("record not found")
        
def delete():
    con=myConnection()
    cur=con.cursor()
    mcode=int(input("enter mobile code to delete: "))
    cur.execute("delete from mobile where mcode='%s'" % (mcode))
    con.commit()
    con.close()
    
    if(cur.rowcount>0):
        print("record deleted")
    else:
        print("record not found")
        
def sort():
    con=myConnection()
    cur=con.cursor()
    cur.execute("select * from mobile order by price")
    rec=cur.fetchall()
    for i in rec:
        for j in i:
            print(j,end="\t")
        print()
    con.close()
    
print("1.save\n2.display\n3.search\n4.update\n5.delete\n6.sort")
ch="y"
while(ch.lower()=="y".lower()):
    op=int(input("enter your option from above menu: "))
    if(op==1):
        save()
    elif(op==2):
        display()
    elif(op==3):
        search()
    elif(op==4):
        update()
    elif(op==5):
        delete()
    elif(op==6):
        sort()
    else:
        print("invalid option opted: ")
    ch=input("want to continue? ")
```
